```python
df["price"].fillna(df['price'].median() ,inplace = True)
```

```python
df["sqft_living"].fillna(df['sqft_living'].median() ,inplace = True)
df["sqft_lot"].fillna(df['sqft_lot'].median() ,inplace = True)
df["view"].fillna(df['view'].median() ,inplace = True)
df["sqft_above"].fillna(df['sqft_above'].median() ,inplace = True)
df["sqft_basement"].fillna(df['sqft_basement'].median() ,inplace = True)

```

```python
df["price"].isnull().sum()
df["sqft_lot"].isnull().sum()
df["view"].isnull().sum()
df["sqft_above"].isnull().sum()
df["sqft_basement"].isnull().sum()
```

```python
# Filling numerical columns
rating_median = df['price'].median()
df['price'] = df['price'].fillna(rating_median)
```

```python
(df.isnull().sum() / len(df)) * 100
```

```python
df['city'].mode()
```

```python
df['street'].mode()
```

```python
df["city"] = df["city"].fillna(df['city'].mode()[0])
```

```python
df["waterfront"] = df["waterfront"].fillna(df['waterfront'].mode()[0])
```

```python
df["floors"] = df["floors"].fillna(df['floors'].mode()[0])
```

```python
df["condition"] = df["condition"].fillna(df['condition'].mode()[0])
```

```python
df["yr_built"] = df["yr_built"].fillna(df['yr_built'].mode()[0])
```

```python
df["street"] = df["street"].fillna(df['street'].mode()[0])
```

```python
df["date"] = df["date"].fillna(df['date'].mode()[0])
```

```python
df["city"].isnull().sum()
df["waterfront"].isnull().sum()
df["floors"].isnull().sum()
df["condition"].isnull().sum()
df["yr_built"].isnull().sum()
df["street"].isnull().sum()
df["date"].isnull().sum()
```

```python
#line plot
plt.figure(figsize = (25,35))
plt.plot(df["bedrooms"])
plt.title("housing",fontsize = 15)
plt.ylabel("bedrooms",fontsize = 15)
plt.xlabel("bathrooms", fontsize = 15)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Bar plot
plt.figure(figsize = (25,25))
plt.bar(df["floors"],df["city"])

plt.ylabel("city",fontsize = 15)
plt.xlabel("floors", fontsize = 15)
plt.xticks(rotation = 90)
plt.grid()

plt.savefig("Bar_housing")
plt.show()
```

```python
#Bar plot
plt.figure(figsize = (25,25))
plt.bar(df["street"],df["city"])

plt.ylabel("city",fontsize = 15)
plt.xlabel("street", fontsize = 15)
plt.xticks(rotation = 90)
plt.grid()

plt.savefig("Bar_housing")
plt.show()
```

```python
#Scatter plot
plt.figure(figsize = (30,15))
plt.scatter(df["city"],df["street"])
plt.title("housing",fontsize = 15)
plt.ylabel("street",fontsize = 15)
plt.xlabel("city", fontsize = 15)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Box plot
plt.figure(figsize = (35,15))
plt.boxplot(df["price"], vert = False)
plt.title("housing",fontsize = 15)
plt.ylabel("price",fontsize = 15)
plt.xlabel("Index", fontsize = 15)

plt.show()
```

```python
#Box plot
plt.figure(figsize = (35,15))
plt.boxplot(df["bedrooms"], vert = False)
plt.title("housing",fontsize = 15)
plt.ylabel("bedrooms",fontsize = 15)
plt.xlabel("Index", fontsize = 15)

plt.show()
```
